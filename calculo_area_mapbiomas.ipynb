```python
#Bibliotecas
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as cl
```

```python
# Localização dos arquivos
mapbiomas2020 = '/content/drive/MyDrive/dados_doutorado/mapbiomas_2020.gpkg'
```

```python
gpk_mapbiomas = gpd.read_file(mapbiomas2020, layer = 'mapbiomas_2020_albers_sirgas2000')
#gpk_mapbiomas.head()
```

```python
gpk_mapbiomas.crs
```

```python
print(gpk_mapbiomas.keys())
```

```python
#Selecionar colunas principais
data_2020 = gpk_mapbiomas[['class_2020', 'classes_geral', 'area_km2', 'geometry']]
data_2020.columns
```

```python
#Número de linhas
print('Linhas: ', len(data_2020['class_2020']))
#Número de classes
print('Classes: ', data_2020['class_2020'].nunique())
# Número de classes agrupadas
print('Classes agrupadas: ', data_2020['classes_geral'].nunique())
```

```python
classes_detalhadas = data_2020.groupby('class_2020')[['area_km2']].sum(numeric_only=True)
classes_detalhadas
```

```python
# Informação importante para escrever no texto
classes_generalizadas = data_2020.groupby('classes_geral')[['area_km2']].sum(numeric_only=True)
classes_generalizadas
```
